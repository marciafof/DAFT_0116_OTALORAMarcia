#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.

````python
# This is the url you will scrape in this exercise
url ='https://en.wikipedia.org/wiki/Python' 
html = requests.get(url).text
document = bs4.BeautifulSoup(html, "html.parser")
print(document.prettify())
````

````python
# Content bloack
contentblock = document.find("div", id="bodyContent").find("div",class_="mw-body-content mw-content-ltr",dir="ltr", id="mw-content-text", lang="en")
# finding parent <ul> tag
allinblocks = contentblock.find_all("ul")

# finding all li tags in ul and printing the text within it
print("This is an inner link list within wiki")
list_links =[]
for data1 in allinblocks:
    for li in data1.find_all("li"):
        print(li.find("a")["href"])
        list_links.append(li.find("a")["href"])
````

####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
html = requests.get(url).text
document = bs4.BeautifulSoup(html, "html.parser")
print(document.prettify())
````

````python
# Content bloack
contentblock = document.find("body").find("div",id="content").find("tbody",id="tbody")
#print(contentblock)
# finding parent <ul> tag#
allinblocks = contentblock.find_all("tr")
#print(allinblocks)
````

````python
####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.
""" list_of_lists = []
list_of_lists.append([1,2,3])
list_of_lists.append([4,5,6])

pd.DataFrame(list_of_lists, columns=['A', 'B', 'C']) """
import pandas as pd

list_eqs = []


for blocki in allinblocks:
    datetimei = blocki.find("td", class_="tabev6").find("a").text.replace(u'\xa0', u' ')
    datei = datetimei.split(" ")[0]; print(datei)
    timei = datetimei.split(" ")[-1]; print(timei)
    coords = blocki.find_all("td", class_="tabev1")
    zone = blocki.find_all("td", class_="tabev2")
    lati =  coords[0].text.strip() + zone[0].text
    loni = coords[1].text.strip() + zone[1].text
    regioni =  blocki.find("td", class_="tb_region").text
    eventi = [datei,timei, lati, loni, regioni]
    list_eqs.append(eventi)
    if len(list_eqs) >20:
        break
````

````python
df_eqs = pd.DataFrame(list_eqs, columns=["date", "time", "latitude", "longitude", "region name"]) 
df_eqs
````

#### Display the top 10 languages by number of native speakers stored in a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
html = requests.get(url).text
document = bs4.BeautifulSoup(html, "html.parser")
print(document.prettify())
````

````python
# Content bloack
contentblock = document.find("div", id="mw-content-text").find("table")
#print(contentblock)
# finding parent <ul> tag#
allinblocks = contentblock.find("tbody").find_all("tr")
print(allinblocks)
````

````python
list_lang = []
for tri in allinblocks:
    #print(tri)
    info = tri.find_all("td")
    cnumber = 0
    for column in info:
        if cnumber == 0 and column.find("a").text:
            cnumber +=1
            name_lang = column.find("a").text
           # print(name_lang)
        elif cnumber ==1:
            num_pop = float(column.text.strip())
#            print(num_pop)
            cnumber +=1
        if cnumber >=2:
            list_lang.append([name_lang,num_pop])
            break
df_lang = pd.DataFrame(list_lang, columns=["Language", "Native speakers \n in millions"])
df_lang.head()
````
