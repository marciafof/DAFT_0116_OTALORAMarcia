#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fitbit["Minutes Lightly Active"]
```

```python
def create_scatterplot(ax, df, xcolumn, ycolumn, color="b", markertype="o"):
    ax.scatter(df[xcolumn],df[ycolumn], c=color, marker=markertype)
    ax.set_xlabel(xcolumn)
    ax.set_ylabel(ycolumn)
    return ax
```

```python
# your code here
fig, axes = plt.subplots(1,3, figsize=(12,4), sharey=True)
for axi in range(3):
    ax = axes.flatten()[axi]
    if axi == 0:
        create_scatterplot(ax, fitbit, "Minutes Lightly Active", "Steps", color="gold", markertype="o")
    elif axi == 1:
        create_scatterplot(ax, fitbit, "Minutes Very Active", "Steps", color="green", markertype="o")
    else:
        create_scatterplot(ax, fitbit, "Minutes Sedentary", "Steps", color="red", markertype="o")
fig.savefig("figures/scatter_minactivity_steps.png", dpi=300)
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
def create_histogram(ax, df, columndata, nbins=10,color="b", label=None):
    ax.hist(df[columndata],bins=nbins, color=color, label=label)
    return ax
```

```python
# your code here
fig, axes = plt.subplots(1,3)
for iax, nbins in enumerate([10,20,50]):
    ax = axes.flatten()[iax]
    create_histogram(ax, titanic, "Age", nbins=nbins,color="b",label=f"N bins: {nbins}")
    ax.legend()

```

#### How does the bin size affect your plot?

The bin size can result in a distribution looking more or less "noisy". This inffluences mostly the perception of the person studying the distribution since it modifies its shape. The higher the number of bins at the end we can get closer to a dot plot where each individual result is plotted as a single point. However, it can also create the opposite effect if there is no a lot of data. Hence, choosing the right number of bins of a histogram is necessary to visualize the actual distribution of a numeric variable.
