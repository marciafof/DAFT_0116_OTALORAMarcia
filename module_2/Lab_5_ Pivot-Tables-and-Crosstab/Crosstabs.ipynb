### 2. Use crosstabs to show the number of moovies per country

```python
pd.crosstab(df.country,df.movie_title)
```

```python
#pd.crosstab(df.country,df.color, values = df.movie_title, aggfunc=np.count_nonzero, margins=True,margins_name="Total").iloc[:,[ -1]].sort_values("Total", ascending=False)
pd.crosstab(df.country,df.movie_title,margins=True,margins_name="Total").iloc[:,[ -1]].sort_values("Total", ascending=False) #By default is COUNT!
```

### 3. Create a crosstab to show the number of films per director and genre splitted by country

```python
pd.crosstab(df.director_name, [df.country,df.genres], margins=True, margins_name="Total")
```

### 4. Average duration of filmes grouped by color

```python
pd.crosstab(df.genres, df.color, values=df.duration, aggfunc=np.mean)
```

### 5. Average number of likes per director (assuming that total film's likes are counted as a sum of actor's likes)

```python
#'actor_3_facebook_likes'
#'actor_1_facebook_likes'
#'actor_2_facebook_likes'
df["new_director_nlikes"] = df.actor_1_facebook_likes+df.actor_2_facebook_likes+df.actor_3_facebook_likes
pd.crosstab(df.director_name, df.new_director_nlikes, values=df.new_director_nlikes, aggfunc=np.nanmean, margins=True,margins_name="Total").loc[:,["Total"]]
```
